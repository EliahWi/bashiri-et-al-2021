```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0.1, 10, 500)
p_x = stats.gamma(2, 0, 1.).pdf(x)
```

```python
fig, ax = plt.subplots(figsize=(4, 2), dpi=250)

lw = 2
zero_threshold = .2

# gamma
ax.plot(x + zero_threshold - .05, p_x, lw=lw, c='firebrick')
ax.fill_between(x + zero_threshold - .055, p_x, color='firebrick', alpha=.4, edgecolor="none")

# uniform
ax.plot([0, 0], [0, .5], c="firebrick", lw=lw)
ax.plot([zero_threshold, zero_threshold], [p_x[0], .5], c="firebrick", lw=lw)
ax.plot([0, zero_threshold], [.5, .5], c="firebrick", lw=lw)
ax.fill_between([0, zero_threshold + .05], [.5, .5], color='firebrick', alpha=.4, edgecolor="none")

ax.set(xlim=(-.1, 8), xticks=[], yticks=[])
sns.despine(left=True, top=True)

fig.savefig("single_model_sketch.png", bbox_inches="tight")
```
