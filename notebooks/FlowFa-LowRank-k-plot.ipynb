```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,20))
ks = [5, 10, 50, 100, 200, 500]
for file in (os.listdir("./k-finder-results")):
    data = np.load(f"./k-finder-results/{file}")
    if "loss" in file:
        axe = 0
        ax[axe].set_ylabel("log likelihood")
        ax[axe].plot(ks, data/1000/np.log(2), marker="*", linewidth=1)
    else: 
        axe = 1
        ax[axe].set_ylabel("time in s")
        ax[axe].plot(ks, data, marker="*", linewidth=1)
        
    
    ax[axe].set_xlabel("k")
    
ax[0].set_xticks(ks)
ax[1].set_xticks(ks)
plt.show()
fig.savefig("images/k-finder.pdf", format="pdf",dpi=300)
```
